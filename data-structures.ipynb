### Float

A Python float is a numberical data type that represents a floating-point number.A floating-pointer number is a number with a decimal point or exponent notation, indicating that a number is a certain number of digits before and after the decimal point or exponent.

##### Creating floating pointer number

````python
f = 1.46
f = 1.45e3
````

##### Round, floor and ceil

````python
f = 1.4567
print(round(f))
print(round(f, 2))
````

````python
from math import floor, ceil

x = floor(1.23)
y = ceil(1.67)

print(x, y)
````

##### Python Tuple VS List

- Tuple is immutable but list is mutable.
- Tuple has better on performance than list, because tuple uses a simple array-like structure in implimentation, however list is implimented as a linked list with operation like inserting elements at random position.

##### Python Tuple Vs Set

- Tuple can hold duplicates while a set can't.

  The entire purpose of a set is its inability of containing duplicates. it's an excellent tool for deduplicating your data.

##### Converting Tuple to Set and List

````python
t = 1, 2, 3, 4
print('convert a tuple to list type using list function', list(t))
print('convert a tuple to list type using * operation', [*t])
````

````python
t = 2, 4, 6, 8
print('convert a tuple to set type using set constructor', set(t))
print('convert a tuple to set type using * operation', {*t})
````

##### adding and removing elements

````python
my_list = [1, 2]
my_list.append('a')
print(my_list)

l1 = [1, 2]
l3 = l1 + my_list
print(l3)

l2 = [7, 7]
l3.extend(l2)
print(l3)

l3.pop()
print(l3)

l3.remove(2)
print(l3)

del(l3[2])
print(l3)

deduplicated = list(set(l3))
print(deduplicated)

deduplicated[0] = 100
print(deduplicated)

deduplicated.clear()
print(deduplicated)
````

### Set

Python sets are unordered collections of distinct objects.

##### how to create a python set

````python
name = { "Eric", "Ali", "John" }

mixed_set = { 'a', 1, (1,2) }

print(name, mixed_set)

my_set = set()
my_set.add(1)
my_set.add('Erik')
my_set.update(range(3))
my_set.update(['a', 'b'])
print(my_set)
print(set([1, 2, 3]))
print(set(range(3)))

my_set_comprehension = { x for x in 'Hi, my name is...' if x not in '., '}
print(my_set_comprehension)
````

##### Mathematical Python set operations
This section will demonstrate and explain all the mathematical set operations. 

###### Find the difference between Python sets

````python
A = { 1, 2, 3, 4, 5 }
B = { 3, 4, 5, 6, 7 } 

print(A - B)
print(B - A)
````

###### Find the symmetric difference between Python sets

````python
A = { 1, 2, 3, 4, 5 }
B = { 3, 4, 5, 6, 7 }

print(A ^ B)
````

###### Find the intersection of two Python sets

````python
A = { 1, 2, 3, 4, 5 }
B = { 3, 4, 5, 6, 7 }

print(A & B)
````

###### Union

````python
A = { 1, 2, 3 }
B = { 3, 4, 5 }

print(A | B)
````

###### Subsets and supersets

````python
A = { 1, 2, 3 }
B = { 1, 2, 3, 4, 5 }
C = { 1, 2 ,3, 10 }

print(A < B)
print(C < B)
print(B > A)
print(B > C)
print(A < A)
print(A <= A)
print(A >= A)
````

### Dictionary

The Python dictionary is one of the language's most powerful data types. In other programming languages and computer science in general, dictionaries are also known as associative arrays. They allow you to associate one or more keys to values.

##### Creating a Python Dictionary

````python
phone_numbers = { 'Jack': '070-02222748', 'Pete': '010-2488634' }
my_empty_dict = { }

print(phone_numbers['Jack'])
````

##### Access and delete a key-value pair

````python
phone_numbers = { 'Jack': '070-02222748', 'Pete': '010-2488634' }
phone_numbers['Eric'] = '06-10101010'
del(phone_numbers['Jack'])

print(phone_numbers)

pete = phone_numbers.get('Pete', 80)
print(pete, phone_numbers.get('default'))
````

##### More ways to create a Python dictionary
Depending on your data source, there are more advanced ways to initialize a dictionary that might come in handy.

###### Using the dict() constructor

````python
my_dict = dict([ ('Jack', '070-02222748'), ('Pete', '010-2488634'), ('Eric', '06-10101010')])
print(my_dict)
````

###### Dictionary Comprehensions

````python
my_dict = {x: x**2 for x in (2, 4, 6)}
print(my_dict)
````

###### Using dict.fromkeys

````python
names = ('Eric', 'Martha', 'Ellen')
phone_numbers = dict.fromkeys(names, None)
print(phone_numbers)
````

##### Check if a key exists in a Python dictionary

````python
phone_numbers = { 'Jack': '070-02222748', 'Pete': '010-2488634' }
print('Jack' in phone_numbers, 'Jack' not in phone_numbers)
````

##### Gettting the length of a Python dictionary

````python
phone_numbers = {'Jack': '070-02222748',
                 'Pete': '010-2488634',
                 'Eric': '06-10101010'}
print(len(phone_numbers))                 
````

##### Dictionary view objects
###### dict.keys() and dict.values()

````python
phone_numbers = { 'Jack': '070-02222748',
                  'Pete': '010-2488634',
                  'Eric': '06-10101010'}
names = phone_numbers.keys()
numbers = phone_numbers.values()
phone_numbers['Linda'] = '030-987612312'
print(names)
print(numbers)

for number in numbers:
    print(number)
````

###### dict.items(): loop through a Python dictionary

````python
phone_numbers = dict([('Jack', '070-02222748'), ('Pete', '010-2488634'), ('Eric', '06-10101010')])
for name, phonenr in phone_numbers.items():
    print(name, phonenr)
````

##### More ways to get all the keys

````python
phone_numbers = {
'Jack': '070-02222748', 
'Pete': '010-2488634', 
'Eric': '06-10101010'
}

print(list(phone_numbers))
print(sorted(phone_numbers))
````

##### Merging dictionaries

````python
dict1 = { 'a': 1, 'b': 2 }
dict2 = { 'b': 3, 'c': 4 }

merged = dict1 | dict2
merged2 = { **dict1, **dict2 }
print(merged, merged2)
````

##### Comparing Python dictionaries

````python
first_dict = { 'a': 1, 'b': 2, 'c': 'a string' }
second_dict = { 'a': 1, 'b': 2, 'c': 'a string'}
print(first_dict == second_dict)
````
